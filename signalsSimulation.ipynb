```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0, 10, 1)  


sinal1 = np.sin(0.2 * np.pi * n) + np.random.normal(0, 0.5, len(n))  
sinal2 = np.cos(0.5 * np.pi * n) + np.random.normal(0, 0.7, len(n))  


plt.stem(n, sinal1, label='Sinal 1 (Não Estacionário)')
plt.stem(n, sinal2, label='Sinal 2 (Não Estacionário)')
plt.xlabel('Amostras (n)')
plt.ylabel('Amplitude')
plt.title('Sinais Discretos Não Estacionários Multivariáveis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo o tempo discreto
n = np.arange(0, 10, 1)  # Amostras de 0 a 9

# Gerando sinais discretos não estacionários multivariáveis
sinal1 = np.sin(0.2 * np.pi * n) + np.random.normal(0, 0.5, len(n))  # Sinal discreto com ruído
sinal2 = np.cos(0.5 * np.pi * n) + np.random.normal(0, 0.7, len(n))  # Outro sinal discreto com ruído

# Plotando os sinais discretos
plt.stem(n, sinal1, label='Sinal 1 (Não Estacionário)')
plt.stem(n, sinal2, label='Sinal 2 (Não Estacionário)')
plt.xlabel('Amostras (n)')
plt.ylabel('Amplitude')
plt.title('Sinais Discretos Não Estacionários Multivariáveis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


t = np.linspace(0, 10, 1000)  


sinal1 = np.sin(2 * np.pi * 1 * t) + np.random.normal(0, 0.5, t.size)  
sinal2 = np.cos(2 * np.pi * 2 * t) + np.random.normal(0, 0.7, t.size)  


plt.figure(figsize=(10, 6))
plt.plot(t, sinal1, label='Sinal 1 (Não Estacionário)')
plt.plot(t, sinal2, label='Sinal 2 (Não Estacionário)')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Sinais Não Estacionários Multivariáveis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


n = np.arange(0, 10, 1)  


sinal1 = np.sin(0.2 * np.pi * n) + np.random.normal(0, 0.5, len(n))  
sinal2 = np.cos(0.5 * np.pi * n) + np.random.normal(0, 0.7, len(n))  


plt.style.use('seaborn-darkgrid')

# Plotando os sinais discretos
plt.figure(figsize=(10, 6))
plt.stem(n, sinal1, label='Sinal 1 (Não Estacionário)', linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(n, sinal2, label='Sinal 2 (Não Estacionário)', linefmt='g-', markerfmt='go', basefmt=' ')
plt.xlabel('Amostras (n)')
plt.ylabel('Amplitude')
plt.title('Sinais Discretos Não Estacionários Multivariáveis')
plt.legend()
plt.grid(True)
plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 10, 1000)  


sinal1 = np.sin(2 * np.pi * 1 * t) * np.exp(-0.1 * t)  
sinal2 = np.cos(2 * np.pi * 2 * t) * np.exp(-0.05 * t)  


plt.figure(figsize=(10, 6))
plt.plot(t, sinal1, label='Sinal 1 (Não Estacionário)')
plt.plot(t, sinal2, label='Sinal 2 (Não Estacionário)')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Sinais Contínuos Não Estacionários Multivariáveis')
plt.legend()
plt.grid(True)
plt.show()
```
